```
import folium

center_coords = [10.781981, 106.697413]

m = folium.Map(location=center_coords, zoom_start=16, control_scale=True)

folium.TileLayer('OpenStreetMap').add_to(m)
folium.TileLayer('cartodbpositron', name='Bản đồ sáng').add_to(m)

poi_list = [
    {"name": "UEH - Cơ sở A", "coords": [10.7819, 106.6974], "icon": "graduation-cap", "color": "red"},
    {"name": "Nhà Văn hóa Thanh Niên", "coords": [10.7821, 106.6961], "icon": "users", "color": "blue"},
    {"name": "Bệnh viện Quận 3", "coords": [10.7845, 106.6835], "icon": "plus", "color": "green"},
    {"name": "Diamond Plaza", "coords": [10.7813, 106.6991], "icon": "shopping-cart", "color": "orange"},
    {"name": "Bưu điện Thành phố", "coords": [10.7799, 106.6999], "icon": "envelope", "color": "purple"},
    {"name": "Dinh Độc Lập", "coords": [10.7770, 106.6953], "icon": "university", "color": "cadetblue"}
]

for poi in poi_list:
    folium.Marker(
        location=poi["coords"],
        popup=f"<b>{poi['name']}</b>",
        tooltip="Click để xem chi tiết",
        icon=folium.Icon(color=poi["color"], icon=poi["icon"], prefix='fa')
    ).add_to(m)

folium.LayerControl().add_to(m)

m
```
