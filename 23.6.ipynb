```
import osmnx as ox
import matplotlib.pyplot as plt

# 1. Tải mạng lưới giao thông (ví dụ: Quận 1, TP.HCM)
place_name = "District 1, Ho Chi Minh City, Vietnam"
graph = ox.graph_from_place(place_name, network_type='drive')

# 2. Trực quan hóa
ox.plot_graph(graph, node_size=10, node_color='r')

# 3. Phân tích cơ bản
stats = ox.basic_stats(graph)
print(f"Số lượng nút giao (intersections): {stats['n']}")
print(f"Chiều dài đường trung bình: {stats['edge_length_avg']:.2f} m")
```
