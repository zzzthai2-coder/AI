```
import osmnx as ox
import networkx as nx
import folium

place = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type="drive")

start = (10.7725, 106.6980)
end = (10.7800, 106.7000)

orig = ox.distance.nearest_nodes(G, start[1], start[0])
dest = ox.distance.nearest_nodes(G, end[1], end[0])

route_dijkstra = nx.shortest_path(G, orig, dest, weight="length")

route_astar = nx.astar_path(G, orig, dest, weight="length")

m = folium.Map(location=start, zoom_start=14)

def draw_route(route, color):
    coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]
    folium.PolyLine(coords, color=color, weight=5).add_to(m)

draw_route(route_dijkstra, "blue")
draw_route(route_astar, "red")

folium.Marker(start, popup="Chợ Bến Thành").add_to(m)
folium.Marker(end, popup="Bưu điện Trung tâm Sài Gòn").add_to(m)

m
```
