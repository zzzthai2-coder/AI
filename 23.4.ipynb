```
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
gdf = gpd.read_file("https://raw.githubusercontent.com/TungTh/tungth.github.io/master/data/vn-provinces.json")
col_name = 'Name' if 'Name' in gdf.columns else gdf.columns[0]
danh_sach_tinh = gdf[col_name].tolist()
np.random.seed(42)
doanh_thu_gia_lap = np.random.randint(100, 2000, size=len(danh_sach_tinh))
df_kinh_doanh = pd.DataFrame({'Ten_Khu_Vuc': danh_sach_tinh, 'Doanh_Thu_Ty_VND': doanh_thu_gia_lap})
merged_gdf = gdf.merge(df_kinh_doanh, left_on=col_name, right_on='Ten_Khu_Vuc')
fig, ax = plt.subplots(1, 1, figsize=(10, 12))
merged_gdf.plot(column='Doanh_Thu_Ty_VND',cmap='OrRd', linewidth=0.5, ax=ax, edgecolor='gray',legend=True, legend_kwds={'label': "Doanh thu (Tỷ VNĐ)", 'orientation': "horizontal", 'shrink': 0.7})
ax.set_title("BẢN ĐỒ CHOROPLETH: DOANH THU THEO TỈNH THÀNH", fontdict={'fontsize': 14, 'fontweight': 'bold'})
ax.axis('off')

plt.tight_layout()
plt.show()
```
