```
import folium
from folium.plugins import HeatMap
import numpy as np

center = [10.7819, 106.6974]

data = (np.random.normal(size=(100, 2), loc=center, scale=0.005)).tolist()

m = folium.Map(location=center, zoom_start=15)

HeatMap(data,
        radius=15,
        blur=10,
        min_opacity=0.5,
        gradient={0.4: 'blue', 0.65: 'lime', 1: 'red'}
       ).add_to(m)


title_html = '''
             <div style="position: fixed;
                         bottom: 50px; left: 50px; width: 250px; height: 90px;
                         background-color: white; border:2px solid grey; z-index:9999; font-size:14px;
                         padding: 10px;">
             <b>Bản đồ nhiệt (23.3)</b><br>
             Vùng <span style="color:red">Đỏ</span>: Mật độ đơn hàng cao<br>
             Ý nghĩa: Cần tập trung shipper tại đây.
             </div>
             '''
m.get_root().html.add_child(folium.Element(title_html))

m
```
