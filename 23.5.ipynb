```
import folium

center = [10.7725, 106.6980]
map_235 = folium.Map(location=center, zoom_start=12)

circles = [
    (10000, 'red', '10km'),
    (5000, 'blue', '5km'),
    (3000, 'green', '3km')
]

for r, c, l in circles:
    folium.Circle(
        location=center,
        radius=r,
        color=c,
        fill=True,
        fill_opacity=0.2,
        popup=l
    ).add_to(map_235)

folium.Marker(
    center,
    popup="Chợ Bến Thành",
    tooltip="Chợ Bến Thành"
).add_to(map_235)

map_235
```
