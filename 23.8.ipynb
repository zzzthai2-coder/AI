```


import folium
import random
from geopy.distance import geodesic

center = (10.7725, 106.6980)

customers = [(10.77 + random.uniform(-0.01, 0.01), 106.69 + random.uniform(-0.01, 0.01)) for _ in range(5)]
drivers = [(10.77 + random.uniform(-0.01, 0.01), 106.69 + random.uniform(-0.01, 0.01)) for _ in range(5)]

matches = []
for c in customers:
    nearest = min(drivers, key=lambda d: geodesic(c, d).meters)
    matches.append((c, nearest))

m = folium.Map(location=center, zoom_start=14)

for c in customers:
    folium.Marker(c, icon=folium.Icon(color='blue'), popup="Khách").add_to(m)

for d in drivers:
    folium.Marker(d, icon=folium.Icon(color='green'), popup="Tài xế").add_to(m)

for c, d in matches:
    folium.PolyLine([c, d], color="red").add_to(m)

m
```
