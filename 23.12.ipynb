```
import folium
from geopy.distance import geodesic

tai_xe = [10.7719, 106.6983]
danh_sach_khach = [
    [10.7797, 106.7048],
    [10.7745, 106.7060],
    [10.7667, 106.6950],
    [10.7680, 106.7010]
]

lo_trinh = [tai_xe] + sorted(danh_sach_khach, key=lambda x: geodesic(tai_xe, x).meters)

m = folium.Map(location=tai_xe, zoom_start=15)

folium.Marker(tai_xe, icon=folium.Icon(color='red')).add_to(m)

for k in danh_sach_khach:
    folium.Marker(k, icon=folium.Icon(color='blue')).add_to(m)

folium.PolyLine(lo_trinh, color='red', weight=8).add_to(m)

m
```
